# 1.1. Dataset'in İçeriğini Anlamak ve Veri Setini Yüklemek 🚀

Bu adımda, WDBC (Wisconsin Diagnostic Breast Cancer) veri setini yükleyip genel yapısını anlamaya çalışacağız:
- **Veri Yükleme:** Veri setini yükleyip sütun isimlerini atayacağız. 📂
- **Veri İncelemesi:** Veri setinin ilk 5 satırını görüntüleyerek genel yapısını anlayacağız. 👀
- **Veri Bilgisi:** Sütunların veri tiplerini ve doluluk durumunu kontrol edeceğiz. 🧠
- **Eksik Veri Kontrolü:** Veri setinde eksik veri olup olmadığını kontrol edeceğiz. 🧹;

```python
# Importing necessary libraries
import pandas as pd

# File path and column names for the dataset
file_path = './Data/wdbc.data'
column_names = [
    "ID", "Diagnosis", "Radius_mean", "Texture_mean", "Perimeter_mean",
    "Area_mean", "Smoothness_mean", "Compactness_mean", "Concavity_mean",
    "Concave_points_mean", "Symmetry_mean", "Fractal_dimension_mean",
    "Radius_se", "Texture_se", "Perimeter_se", "Area_se", "Smoothness_se",
    "Compactness_se", "Concavity_se", "Concave_points_se", "Symmetry_se",
    "Fractal_dimension_se", "Radius_worst", "Texture_worst",
    "Perimeter_worst", "Area_worst", "Smoothness_worst",
    "Compactness_worst", "Concavity_worst", "Concave_points_worst",
    "Symmetry_worst", "Fractal_dimension_worst"
]

# Loading the dataset
try:
    data = pd.read_csv(file_path, header=None, names=column_names)
    print(" Dataset successfully loaded.")
except FileNotFoundError:
    print(f"Error: {file_path} not found. Please verify the file path.")
    data = None

# Initial exploration
if data is not None:
    print("\nFirst 5 rows of the dataset:")
    print(data.head())

    print("\nDataset columns:")
    print(data.columns)

    # Checking for missing values
    print("\nMissing values per column:")
    print(data.isnull().sum())

    # Checking unique values in the Diagnosis column
    print("\nUnique values in the 'Diagnosis' column:")
    print(data['Diagnosis'].unique())
```

# 1.2. Veri Setinin Özet İstatistiklerini Çıkarma 📊

Bu adımda veri setinin genel özet istatistiklerini çıkararak analiz için bir temel oluşturacağız:
- **Sütun Bilgileri:** Veri setindeki her bir sütunun veri tipi, doluluk durumu ve boyutu incelenecek. 📝
- **Özet İstatistikler:** Sayısal sütunlar için minimum, maksimum, ortalama ve standart sapma gibi temel istatistiksel bilgiler elde edilecek. 📈

```python
# Displaying dataset information
print("Dataset Information:")
print(data.info())  # Shows data types, non-null values, and memory usage

# Generating summary statistics for numerical columns
print("\nSummary Statistics:")
print(data.describe())  # Provides statistical summary for numerical columns
```

# 1.3. Eksik Veri Kontrolü 🧹

Bu adımda veri setindeki eksik verileri kontrol edeceğiz:
- **Eksik Veri Sayısı:** Her bir sütunda kaç eksik veri olduğunu analiz edeceğiz. 🔍
- **Eksik Veri Yüzdesi:** Veri setindeki eksik verilerin yüzdesini hesaplayacağız. 📊
- **Eksik Verilerin Görselleştirilmesi:** Eksik veri oranlarını bar grafiği ile görselleştireceğiz. 📉

```python
# Checking for missing values
missing_data = data.isnull().sum()
missing_percentage = (missing_data / len(data)) * 100

# Displaying missing data information
print("\nMissing Data per Column:")
print(missing_data)

print("\nMissing Data Percentage per Column:")
print(missing_percentage)

# Visualizing missing data if there are any missing values
if missing_percentage.sum() > 0:
    import matplotlib.pyplot as plt
    import seaborn as sns

    plt.figure(figsize=(12, 6))
    sns.barplot(x=missing_data.index, y=missing_percentage.values, palette="viridis")
    plt.xticks(rotation=45, ha="right")
    plt.title("Missing Data Percentage by Column")
    plt.xlabel("Columns")
    plt.ylabel("Percentage")
    plt.show()
else:
    print("No missing data found in the dataset.")
```

# 1.5. ID Sütununu Kaldırma 🗑️

Bu adımda:
- **Neden ID Kaldırılıyor?** ID sütunu veri analizi için anlamlı bir özellik değildir. Her bir örneğe özgü bir tanımlayıcı olduğu için analitik bağlamda bir katkısı yoktur. 🔗

# 1.6. Diagnosis Sütununu Sayısal Hale Getirme 🔢

Bu adımda:
- **Amaç:** `Diagnosis` (Teşhis) sütununu sayısal bir formata dönüştürerek analiz ve modelleme işlemleri için uygun hale getireceğiz. 👩‍💻
- **Dönüşüm:** Kategorik değerler olan `M` (Malignant - Kötü Huylu) ve `B` (Benign - İyi Huylu) sırasıyla `1` ve `0` değerlerine dönüştürülecek. 🔄
- **Sonuç:** Bu dönüşüm, makine öğrenimi algoritmaları ve sayısal analizler için veri setini daha kullanılabilir hale getirir. 🚀

```python
# Converting the Diagnosis column to numerical format
data['Diagnosis'] = data['Diagnosis'].map({'M': 1, 'B': 0})

# Verifying the conversion
print("Unique values in the 'Diagnosis' column after conversion:")
print(data['Diagnosis'].unique())
```

# 1.8. Korelasyon Matrisi 🔗

Bu adımda:
- **Amaç:** Özellikler arasındaki ilişkileri ve korelasyonları analiz ederek veri seti hakkında derinlemesine bilgi edinmek. 📊
- **Korelasyon Matrisi:** Özellikler arasındaki korelasyonları analiz edeceğiz. Korelasyon katsayısı `-1` ile `1` arasında değişir:
  - `1` → Güçlü pozitif korelasyon 🤝
  - `0` → Hiçbir korelasyon yok ❌
  - `-1` → Güçlü negatif korelasyon 💔
- **Isı Haritası:** Korelasyon matrisini görselleştirmek için ısı haritası oluşturacağız. 🌡️

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure that the 'Diagnosis' column is numeric (convert 'M' and 'B' to 1 and 0 if not already done)
if data['Diagnosis'].dtype == 'object':
    data['Diagnosis'] = data['Diagnosis'].map({'M': 1, 'B': 0})

# Calculating the correlation matrix (only numeric columns)
correlation_matrix = data.corr()

# Displaying the correlation matrix as a heatmap with improved readability
plt.figure(figsize=(16, 12))  # Adjust figure size for better visibility
sns.heatmap(
    correlation_matrix,
    annot=True,              # Display correlation coefficients
    fmt=".2f",               # Format numbers with two decimal places
    cmap="coolwarm",         # Use a color map for better visual distinction
    cbar=True,               # Include color bar
    square=True,             # Keep the cells square-shaped
    linewidths=0.5,          # Add space between cells
    annot_kws={"size": 8}    # Set font size for cell annotations
)
plt.title("Correlation Matrix Heatmap", fontsize=18)  # Add a title to the plot
plt.xticks(fontsize=10, rotation=45, ha="right")      # Rotate and format x-axis labels
plt.yticks(fontsize=10)                              # Format y-axis labels
plt.tight_layout()                                   # Ensure layout fits within the figure
plt.show()
```

# 1.9. Özelliklerin Normalizasyonu 📏

## Bu adımda:  
- **Amaç:** Veri setindeki özelliklerin farklı ölçeklerden gelmesi durumunda, modelleme sürecinin bu farklılıklardan etkilenmemesi için özelliklerin ölçeklerini standart hale getireceğiz.  
- **Yöntem:** Normalizasyon veya standartlaştırma işlemleri kullanılarak tüm sayısal özellikler aynı ölçeğe getirilecek.  
- **Sonuç:** Model performansını artırarak daha etkili bir öğrenme süreci sağlanır. 🚀  

```python
# Importing StandardScaler for feature normalization
from sklearn.preprocessing import StandardScaler

# Selecting numerical columns to normalize
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Initializing the StandardScaler
scaler = StandardScaler()

# Normalizing the data
data[numeric_columns] = scaler.fit_transform(data[numeric_columns])

# Displaying the first few rows of the normalized dataset
print("Normalized dataset:")
print(data.head())
```

# 1.10. Özellik Seçimi 🧐

## Bu adımda:
- **Amaç:** Modelin doğruluğunu artırmak ve eğitim süresini azaltmak için sadece en önemli özellikleri seçmek.
- **Yöntem:** Özelliklerin model performansı üzerindeki etkisini analiz ederek gereksiz veya az katkı sağlayan özellikleri eleyeceğiz.
- **Araçlar:** Korelasyon matrisi ve belirli algoritmalar kullanılarak en etkili özellikler belirlenecek.
- **Sonuç:** Model daha hızlı ve daha doğru bir şekilde çalışacaktır. 🚀

```python
# Importing SelectKBest and f_classif for feature selection
from sklearn.feature_selection import SelectKBest, f_classif

# Separating the target variable and features
X = data.drop(columns=["Diagnosis"])
y = data["Diagnosis"]

# Applying SelectKBest to extract top 10 features
selector = SelectKBest(score_func=f_classif, k=10)
X_selected = selector.fit_transform(X, y)

# Getting selected feature names
selected_features = X.columns[selector.get_support()]

# Displaying selected features
print("Top 10 selected features based on ANOVA F-statistic:")
print(selected_features)
```

# **1.11. Makine Öğrenimi Modelleri Eğitimi ve Test Edilmesi 🧠**

Bu adımda seçilen özellikler ile makine öğrenimi algoritmaları kullanılarak veri seti üzerinde model eğitimi yapılacaktır:

- Eğitim ve test setleri %70 ve %30 olarak bölünecek. 📊
- Farklı algoritmalar denenerek doğruluk gibi metriklerle değerlendirilecektir.

```python
# Importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Ensure target variable (y) is categorical
if y.dtype == 'float64' or y.dtype == 'int64':
    # Convert continuous target to categorical if necessary
    if len(y.unique()) > 10:  # Assuming >10 unique values means continuous
        print("Converting continuous target variable to categorical classes.")
        import numpy as np
        y = np.digitize(y, bins=np.linspace(y.min(), y.max(), num=4))  # Bin into 3 categories (example)
    else:
        y = y.astype(int)  # Convert to integer for classification models

# Verify the target variable is suitable for classification
print("Unique values in target variable (y):", y.unique())

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# Model 1: Logistic Regression
log_reg = LogisticRegression(random_state=42, max_iter=1000)  # Initialize Logistic Regression
log_reg.fit(X_train, y_train)  # Train the model
y_pred_log_reg = log_reg.predict(X_test)  # Predict on test data

# Evaluate Logistic Regression
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
print(f"Logistic Regression Accuracy: {accuracy_log_reg:.2f}")
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_log_reg))

# Model 2: Random Forest Classifier
rf_clf = RandomForestClassifier(random_state=42, n_estimators=100)  # Initialize Random Forest
rf_clf.fit(X_train, y_train)  # Train the model
y_pred_rf = rf_clf.predict(X_test)  # Predict on test data

# Evaluate Random Forest Classifier
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.2f}")
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

# **1.12. Modellerin Değerlendirilmesi 🔍**

Bu adımda Logistic Regression ve Random Forest modellerinin performansı şu yöntemlerle analiz edilecek:

- **Confusion Matrix Görselleştirme:** Sınıflandırma doğruluğu görselleştirilecek. 📊
- **ROC Eğrisi ve AUC Skoru:** Her iki modelin ROC eğrileri çizilecek ve AUC değerleri karşılaştırılacak. 📈
- **Sonuçların Tartışılması:** Hangi modelin daha iyi performans gösterdiği üzerinde durulacak. 🤔

```python
# Importing necessary libraries for evaluation
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# --- Confusion Matrix Visualization ---
# Logistic Regression Confusion Matrix
cm_log_reg = confusion_matrix(y_test, y_pred_log_reg)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_log_reg, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Random Forest Confusion Matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Oranges', cbar=False)
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# --- ROC Curve and AUC ---
# Logistic Regression ROC Curve
fpr_log_reg, tpr_log_reg, _ = roc_curve(y_test, log_reg.predict_proba(X_test)[:, 1])
auc_log_reg = roc_auc_score(y_test, log_reg.predict_proba(X_test)[:, 1])

# Random Forest ROC Curve
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_clf.predict_proba(X_test)[:, 1])
auc_rf = roc_auc_score(y_test, rf_clf.predict_proba(X_test)[:, 1])

# Plotting ROC Curves
plt.figure(figsize=(10, 8))
plt.plot(fpr_log_reg, tpr_log_reg, label=f"Logistic Regression (AUC = {auc_log_reg:.2f})", linestyle='--')
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {auc_rf:.2f})", linestyle='-')
plt.plot([0, 1], [0, 1], 'k--', label="Random Chance")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

# **1.13. Model Sonuçlarının Görselleştirilmesi ve Karşılaştırılması 📊**

Bu adımda, Logistic Regression ve Random Forest modellerinin performansı şu metriklerle karşılaştırılacak:

- **Confusion Matrix Görselleştirmesi**: Doğru ve yanlış sınıflandırmaların sayısını karşılaştıracağız. 📊
- **Precision, Recall ve F1-Score Karşılaştırması**: Hangi modelin hangi metrikte daha iyi olduğunu tartışacağız. 📈

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate metrics for Logistic Regression
precision_log_reg = precision_score(y_test, y_pred_log_reg)
recall_log_reg = recall_score(y_test, y_pred_log_reg)
f1_log_reg = f1_score(y_test, y_pred_log_reg)

# Calculate metrics for Random Forest
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

# Create a comparison table and display it
import pandas as pd

# Precision, Recall, F1-Score for Logistic Regression and Random Forest
comparison = {
    "Model": ["Logistic Regression", "Random Forest"],
    "Precision": [precision_log_reg, precision_rf],
    "Recall": [recall_log_reg, recall_rf],
    "F1-Score": [f1_log_reg, f1_rf]
}

comparison_df = pd.DataFrame(comparison)

# Display the comparison dataframe
print(comparison_df)

# Optionally, you can also visualize this comparison
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))
comparison_df.set_index('Model').plot(kind='bar', ax=ax)
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.legend(title='Metric', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Bu adımda, her iki modelin performansı değerlendirilecek:

- **Logistic Regression** ve **Random Forest** karşılaştırılarak hangi modelin daha iyi sonuç verdiği tartışılacak. 📊

```python
# Final evaluation - Comparing the models
print("Logistic Regression Model Performance:")
print(f"Precision: {precision_log_reg:.2f}")
print(f"Recall: {recall_log_reg:.2f}")
print(f"F1-Score: {f1_log_reg:.2f}")
print("\n")

print("Random Forest Model Performance:")
print(f"Precision: {precision_rf:.2f}")
print(f"Recall: {recall_rf:.2f}")
print(f"F1-Score: {f1_rf:.2f}")

# Suggesting potential improvements for model performance
improvement_suggestions = """
Model Improvements Suggestions:
1. Hyperparameter Tuning:
   - For Logistic Regression: Try using different solvers, regularization techniques (e.g., L1/L2 regularization), and adjusting the max_iter parameter.
   - For Random Forest: Tune parameters such as n_estimators, max_depth, min_samples_split, etc.

2. Cross-validation:
   - Use k-fold cross-validation to get more reliable performance estimates and reduce overfitting.

3. Feature Engineering:
   - Explore feature scaling, feature selection, or adding polynomial features to improve model performance.

4. Advanced Models:
   - Explore using more complex models such as XGBoost or LightGBM for better handling of imbalanced datasets or complex patterns.
"""

print(improvement_suggestions)
```
