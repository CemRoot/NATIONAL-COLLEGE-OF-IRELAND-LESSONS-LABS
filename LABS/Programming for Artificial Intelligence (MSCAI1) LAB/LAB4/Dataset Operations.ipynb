# IRIS Dataset Operations

The Iris dataset is one of the most famous and widely used datasets in machine learning and statistics. It was first introduced by the British biologist and statistician Ronald A. Fisher in 1936 in his paper The use of multiple measurements in taxonomic problems. The dataset is sometimes referred to as the Fisher's Iris dataset or simply the Fisher Iris dataset.

<b>Overview</b> 

The Iris dataset consists of 150 samples from each of three species of iris flowers:

Iris Setosa (label 0)

Iris Versicolor (label 1)

Iris Virginica (label 2)


Each sample has four features (measurements of the flowers in centimeters):


Sepal length

Sepal width

Petal length

Petal width

The goal is often to classify these flowers into their respective species based on the four features using machine learning algorithms. Since it is a well-balanced dataset (50 samples per species), it is frequently used for pattern recognition tasks, data visualization, and demonstrating classification algorithms.

<b>Features and Label</b>

Features: Measurements of the flowers:

Sepal length (cm)

Sepal width (cm)

Petal length (cm)

Petal width (cm)

<b>Label: The species or class of the flower:</b>

0: Setosa

1: Versicolor

2: Virginica


<b>Dataset Structure</b>

The dataset is structured as follows:




<b>Characteristics of the Dataset:</b>

150 total samples: 50 from each class.

4 features per sample (sepal length, sepal width, petal length, petal width).

Balanced classes: Each class has exactly 50 instances, making it a good dataset for classification tasks.

<b>NumPy (Numerical Python):</b>

Purpose: Provides support for arrays, matrices, and a large collection of mathematical functions to operate on these data structures.

Main Use Cases:

Handling large datasets with multi-dimensional arrays.

Performing mathematical and statistical operations.

Efficient handling of large-scale data with vectorized operations.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset into a Pandas DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
sepal_lengths = np.array(df['sepal length (cm)'])
petal_lengths = np.array(df['petal length (cm)'])

# ================================================================
# 1. ARRAY CREATION AND MANIPULATION
# ================================================================

# Reshape array (converting from 1D to 2D)
print("1. Reshaped Sepal Lengths:")
reshaped_sepal_lengths = np.reshape(sepal_lengths, (150, 1))
print(reshaped_sepal_lengths[:5])  # Displaying first 5 rows

# Concatenate arrays
print("\n2. Concatenated Sepal and Petal Lengths:")
concatenated_data = np.concatenate([sepal_lengths, petal_lengths])
print(concatenated_data[:10])  # Display first 10 concatenated elements

# Stack arrays vertically (row-wise)
print("\n3. Vertically Stacked Sepal and Petal Lengths:")
vertical_stack = np.vstack([sepal_lengths, petal_lengths])
print(vertical_stack[:, :5])  # Display first 5 columns of stacked data

# Stack arrays horizontally (column-wise)
print("\n4. Horizontally Stacked Sepal and Petal Lengths:")
horizontal_stack = np.hstack([sepal_lengths.reshape(-1, 1), petal_lengths.reshape(-1, 1)])
print(horizontal_stack[:5])  # Display first 5 rows

# Split the data into multiple sub-arrays
print("\n5. Split Data into 3 Parts:")
split_data = np.split(df.values, 3)
print(split_data[0][:5])  # Display first 5 rows of the first split

# Transpose the array (flip rows and columns)
print("\n6. Transposed Data:")
transposed_data = np.transpose(df.values)
print(transposed_data[:, :5])  # Display first 5 columns of transposed data

# ================================================================
# 2. STATISTICAL FUNCTIONS
# ================================================================

# Mean of sepal length
print("\n7. Mean of Sepal Lengths:", np.mean(sepal_lengths))

# Median of sepal length
print("8. Median of Sepal Lengths:", np.median(sepal_lengths))

# Standard deviation of sepal length
print("9. Standard Deviation of Sepal Lengths:", np.std(sepal_lengths))

# Variance of sepal length
print("10. Variance of Sepal Lengths:", np.var(sepal_lengths))

# Minimum and maximum of sepal length
print("11. Minimum Sepal Length:", np.min(sepal_lengths))
print("12. Maximum Sepal Length:", np.max(sepal_lengths))

# 90th percentile of sepal length
print("13. 90th Percentile of Sepal Lengths:", np.percentile(sepal_lengths, 90))

# Correlation between sepal length and petal length
print("14. Correlation Coefficient between Sepal and Petal Lengths:")
print(np.corrcoef(sepal_lengths, petal_lengths))

# Histogram of sepal lengths
print("\n15. Histogram of Sepal Lengths:")
hist, bin_edges = np.histogram(sepal_lengths, bins=10)
print("Histogram:", hist)
print("Bin Edges:", bin_edges)

# ================================================================
# 3. LINEAR ALGEBRA OPERATIONS
# ================================================================

# Dot product of sepal and petal lengths
print("\n16. Dot Product of Sepal and Petal Lengths:", np.dot(sepal_lengths, petal_lengths))

# Matrix inversion (example matrix)
matrix = np.array([[1, 2], [3, 4]])
inverse_matrix = np.linalg.inv(matrix)
print("17. Inverse of Matrix [[1, 2], [3, 4]]:")
print(inverse_matrix)

# Determinant of the matrix
determinant = np.linalg.det(matrix)
print("18. Determinant of Matrix [[1, 2], [3, 4]]:", determinant)

# Eigenvalues and eigenvectors of the matrix
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print("19. Eigenvalues of Matrix [[1, 2], [3, 4]]:", eigenvalues)
print("20. Eigenvectors of Matrix [[1, 2], [3, 4]]:")
print(eigenvectors)

# Norm (magnitude) of the sepal lengths
print("21. Norm of Sepal Lengths:", np.linalg.norm(sepal_lengths))

# Matrix multiplication of two reshaped arrays
matrix_mult = np.matmul(sepal_lengths.reshape(-1, 1), petal_lengths.reshape(1, -1))
print("22. Matrix Multiplication Result (Sepal Lengths x Petal Lengths):")
print(matrix_mult[:3, :3])  # Display first 3x3 elements of the resulting matrix

# ================================================================
# 4. RANDOM NUMBER GENERATION
# ================================================================

# Generate random numbers between 0 and 1
random_data = np.random.rand(150, 4)
print("\n23. Random Numbers (150x4 Array):")
print(random_data[:5])  # Display first 5 rows

# Generate random integers between 0 and 100
random_ints = np.random.randint(0, 100, size=10)
print("24. Random Integers between 0 and 100:", random_ints)

# Shuffle sepal lengths in place
np.random.shuffle(sepal_lengths)
print("25. Shuffled Sepal Lengths:", sepal_lengths[:5])  # Display first 5 shuffled elements

# Randomly select 5 samples from sepal lengths
random_samples = np.random.choice(sepal_lengths, size=5)
print("26. Randomly Selected Samples from Sepal Lengths:", random_samples)

# ================================================================
# 5. LOGICAL OPERATIONS
# ================================================================

# Find indices where the species is 0 (Setosa)
df['species'] = iris['target']  # Add species to DataFrame
setosa_indices = np.where(df['species'] == 0)
print("\n27. Indices where Species is Setosa (0):", setosa_indices[0][:5])  # Display first 5 indices

# Logical AND: Sepal length > 5 and Petal length < 1.5
condition = np.logical_and(sepal_lengths > 5.0, petal_lengths < 1.5)
print("28. Logical AND (Sepal Length > 5 and Petal Length < 1.5):", condition[:5])

# Logical OR: Sepal length > 5 OR Petal length < 1.5
condition_or = np.logical_or(sepal_lengths > 5.0, petal_lengths < 1.5)
print("29. Logical OR (Sepal Length > 5 or Petal Length < 1.5):", condition_or[:5])
```

<b>Pandas (Python Data Analysis Library):</b>

Purpose: Provides data structures (e.g., DataFrames) to store and manipulate tabular data (think Excel spreadsheets) efficiently.

Main Use Cases:

Handling structured data (rows and columns) with indexing, filtering, and aggregation operations.

Data cleaning, transformation, and merging datasets.

Descriptive statistics and easy access to summary information about datasets.

```python

import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset into a Pandas DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df['species'] = iris['target']  # Add species to the DataFrame

# ================================================================
# 1. DATAFRAME CREATION AND BASIC INSPECTION
# ================================================================

# Display the first 5 rows of the dataframe
print("1. First 5 rows of the dataset:")
print(df.head())

# Display the last 5 rows of the dataframe
print("\n2. Last 5 rows of the dataset:")
print(df.tail())

# Display the column names of the dataframe
print("\n3. Column names:")
print(df.columns)

# Get a summary of the dataframe (non-null counts and data types)
print("\n4. Dataframe info:")
print(df.info())

# Get statistical summary of numerical columns
print("\n5. Statistical summary of numerical columns:")
print(df.describe())

# Get the shape of the dataframe (rows, columns)
print("\n6. Shape of the dataframe (rows, columns):", df.shape)

# ================================================================
# 2. COLUMN AND ROW OPERATIONS
# ================================================================

# Access a single column (sepal length)
print("\n7. Sepal Length column:")
print(df['sepal length (cm)'].head())

# Access multiple columns (sepal length and petal length)
print("\n8. Sepal Length and Petal Length columns:")
print(df[['sepal length (cm)', 'petal length (cm)']].head())

# Filter rows based on a condition (sepal length > 5.0)
print("\n9. Filter rows where Sepal Length > 5.0:")
print(df[df['sepal length (cm)'] > 5.0].head())

# Apply a custom function to transform data (increase sepal length by 10%)
print("\n10. Increase Sepal Length by 10% (first 5 rows):")
df['sepal length (cm)'] = df['sepal length (cm)'].apply(lambda x: x * 1.10)
print(df[['sepal length (cm)']].head())

# Add a new column based on existing data (sepal-petal length ratio)
df['sepal_petal_length_ratio'] = df['sepal length (cm)'] / df['petal length (cm)']
print("\n11. New column 'sepal_petal_length_ratio' (first 5 rows):")
print(df[['sepal length (cm)', 'petal length (cm)', 'sepal_petal_length_ratio']].head())

# Drop a column (sepal_petal_length_ratio)
df = df.drop(columns=['sepal_petal_length_ratio'])
print("\n12. Dropped 'sepal_petal_length_ratio' column (first 5 rows):")
print(df.head())

# ================================================================
# 3. GROUPING AND AGGREGATION
# ================================================================

# Group data by species and compute the mean for each species
print("\n13. Mean of each feature grouped by species:")
print(df.groupby('species').mean())

# Count the number of occurrences for each species
print("\n14. Count of each species:")
print(df['species'].value_counts())

# Aggregate multiple statistics (mean, max) for each species
print("\n15. Aggregate statistics (mean, max) for each species:")
print(df.groupby('species').agg(['mean', 'max']))

# ================================================================
# 4. SORTING AND INDEXING
# ================================================================

# Sort the dataframe by sepal length (ascending)
print("\n16. Data sorted by Sepal Length (ascending):")
print(df.sort_values(by='sepal length (cm)').head())

# Sort the dataframe by petal length (descending)
print("\n17. Data sorted by Petal Length (descending):")
print(df.sort_values(by='petal length (cm)', ascending=False).head())

# Set 'species' as the index of the dataframe
df_with_index = df.set_index('species')
print("\n18. Dataframe with 'species' as index:")
print(df_with_index.head())

# Reset the index back to default
df_reset_index = df_with_index.reset_index()
print("\n19. Dataframe after resetting the index:")
print(df_reset_index.head())

# ================================================================
# 5. MISSING DATA HANDLING
# ================================================================

# Check for missing values
print("\n20. Check for missing values:")
print(df.isnull().sum())

# Introduce some missing values for demonstration
df.loc[0:5, 'sepal length (cm)'] = None
print("\n21. Dataframe with missing values (first 10 rows):")
print(df.head(10))

# Fill missing values with a specified value (mean of the column)
df['sepal length (cm)'].fillna(df['sepal length (cm)'].mean(), inplace=True)
print("\n22. Fill missing values with the mean (first 10 rows):")
print(df.head(10))

# ================================================================
# 6. CONCATENATION AND MERGING
# ================================================================

# Concatenate the original dataframe with itself (along rows)
df_concat = pd.concat([df, df], axis=0)
print("\n23. Concatenate the dataframe with itself (along rows, first 5 rows):")
print(df_concat.head())

# Concatenate along columns
df_concat_cols = pd.concat([df[['sepal length (cm)', 'sepal width (cm)']],
                            df[['petal length (cm)', 'petal width (cm)']]], axis=1)
print("\n24. Concatenate along columns (first 5 rows):")
print(df_concat_cols.head())

# ================================================================
# 7. ADVANCED DATAFRAME OPERATIONS
# ================================================================

# Create a pivot table (average of features for each species)
print("\n25. Pivot table with species as index and average of features:")
pivot_table = df.pivot_table(values=['sepal length (cm)', 'petal length (cm)'],
                             index='species', aggfunc='mean')
print(pivot_table)

# Apply a rolling window operation (e.g., 3-row rolling mean)
print("\n26. Rolling mean for Sepal Length (window size = 3):")
print(df['sepal length (cm)'].rolling(window=3).mean().head(10))

# ================================================================
# 8. EXPORTING AND IMPORTING DATA
# ================================================================

# Export the dataframe to a CSV file
df.to_csv('iris_exported.csv', index=False)
print("\n27. DataFrame exported to 'iris_exported.csv'")

# Read the CSV file back into a DataFrame
df_imported = pd.read_csv('iris_exported.csv')
print("\n28. DataFrame imported from 'iris_exported.csv' (first 5 rows):")
print(df_imported.head())

```

<b>Matplotlib (Mathematical Plotting Library):</b>

Purpose: Provides tools for creating static, animated, and interactive visualizations in Python.

Main Use Cases:

Creating various types of plots: line plots, scatter plots, histograms, bar charts, pie charts, etc.

Customizing visual elements like titles, axes labels, colors, and more.

Making scientific or publication-quality visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset into a Pandas DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df['species'] = iris['species'] = iris['target']  # Add species to the DataFrame

# ================================================================
# 1. BASIC PLOTTING
# ================================================================

# Line plot of Sepal Length vs Index
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length', color='b')
plt.title('Line Plot: Sepal Length vs Index')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.show()

# ================================================================
# 2. SCATTER PLOT
# ================================================================

# Scatter plot of Sepal Length vs Petal Length
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], c='g', label='Data Points')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()

# ================================================================
# 3. HISTOGRAM
# ================================================================

# Histogram of Sepal Length
plt.figure(figsize=(10, 6))
plt.hist(df['sepal length (cm)'], bins=20, color='purple', edgecolor='black')
plt.title('Histogram: Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# ================================================================
# 4. BAR PLOT
# ================================================================

# Bar plot of the mean Sepal Length for each species
mean_sepal_length = df.groupby('species')['sepal length (cm)'].mean()
species_labels = ['Setosa', 'Versicolor', 'Virginica']

plt.figure(figsize=(10, 6))
plt.bar(species_labels, mean_sepal_length, color=['red', 'blue', 'green'])
plt.title('Bar Plot: Mean Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length (cm)')
plt.show()

# ================================================================
# 5. BOX PLOT
# ================================================================

# Box plot for Sepal Length, Petal Length, and Sepal Width
plt.figure(figsize=(10, 6))
df[['sepal length (cm)', 'petal length (cm)', 'sepal width (cm)']].plot(kind='box', grid=True)
plt.title('Box Plot: Sepal and Petal Length/Width')
plt.ylabel('Length/Width (cm)')
plt.show()

# ================================================================
# 6. PIE CHART
# ================================================================

# Pie chart for the proportion of species in the dataset
species_counts = df['species'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'skyblue', 'lightgreen'])
plt.title('Pie Chart: Species Distribution in Iris Dataset')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()

# ================================================================
# 7. PAIR PLOT (with seaborn for more advanced plots)
# ================================================================

import seaborn as sns

# Pair plot showing relationships between features, colored by species
sns.pairplot(df, hue='species', palette='bright')
plt.suptitle('Pair Plot: Features of the Iris Dataset', y=1.02)
plt.show()

# ================================================================
# 8. SUBPLOTS
# ================================================================

# Create subplots for Sepal Length and Sepal Width
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Sepal Length distribution
axs[0].hist(df['sepal length (cm)'], bins=20, color='c', edgecolor='black')
axs[0].set_title('Sepal Length Distribution')
axs[0].set_xlabel('Sepal Length (cm)')
axs[0].set_ylabel('Frequency')

# Sepal Width distribution
axs[1].hist(df['sepal width (cm)'], bins=20, color='m', edgecolor='black')
axs[1].set_title('Sepal Width Distribution')
axs[1].set_xlabel('Sepal Width (cm)')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# ================================================================
# 9. HEATMAP
# ================================================================

# Heatmap of correlation matrix between features
correlation_matrix = df.iloc[:, :-1].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap: Correlation Matrix of Iris Features')
plt.show()

# ================================================================
# 10. 3D SCATTER PLOT
# ================================================================

from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter plot of Sepal Length, Sepal Width, and Petal Length
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['sepal length (cm)'], df['sepal width (cm)'], df['petal length (cm)'],
           c=df['species'], cmap='viridis', s=60)

ax.set_title('3D Scatter Plot: Sepal Length vs Sepal Width vs Petal Length')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')

plt.show()
```
